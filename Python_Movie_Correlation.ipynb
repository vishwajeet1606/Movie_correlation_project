```python
#importing libraries which we will use in the project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r"C:\Users\DELL\Downloads\movies.csv\movies.csv")


```

```python
df.head()
```

```python
# let's see if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# dropping rows with null values
df = df.dropna()
```

```python
# data types for our columns
df.dtypes
```

```python
# change data types of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# create correct year column

df['yearcorrect'] = df['released'].astype(str).str[-20:-15]

df.head()
```

```python
df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df = df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
# drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel("Budget for films")
plt.show()
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel("Movie Features")

plt.show()
```

```python
# Look at company
df.head()
```

```python
# converting the company and other columns to some numeric values to include it in the correlation matrix
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel("Movie Features")

plt.show()
```

```python
df_numerized.corr()
```

```python
# using unstacking

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
